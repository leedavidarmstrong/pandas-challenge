```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
schools_complete = Path(".", "Resources", "schools_complete.csv")
students_complete = Path(".", "Resources", "students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
schools_complete = pd.read_csv(schools_complete)
students_complete = pd.read_csv(students_complete)

# Combine the data into a single dataset. 
schools_students_complete_combined = pd.merge(schools_complete, students_complete, how="left", on=["school_name", "school_name"])
print(schools_students_complete_combined.head())
```

```python
## Calculate the percentage of students with passing reading score
passing_reading_percentage = (students_complete['reading_score'] >= 50).mean() * 100

## Print the percentage of students with passing reading score
#print("Percentage of Students with Passing Reading Score:", passing_reading_percentage)
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# School Summary
# Calculate the number of students per school
students_per_school = schools_students_complete_combined.groupby('school_name')['Student ID'].count()

# Calculate the total school budget per school
total_budget_per_school = schools_students_complete_combined.groupby('school_name')['budget'].mean()

# Calculate the per student budget per school
per_student_budget_per_school = total_budget_per_school / students_per_school

# Calculate the average math score per school
average_maths_score_per_school = schools_students_complete_combined.groupby('school_name')['maths_score'].mean()

# Calculate the average reading score per school
average_reading_score_per_school = schools_students_complete_combined.groupby('school_name')['reading_score'].mean()

# Calculate the percentage of students passing math per school
passing_maths_per_school = schools_students_complete_combined[schools_students_complete_combined['maths_score'] >= 50].groupby('school_name')['Student ID'].count() / students_per_school * 100

# Calculate the percentage of students passing reading per school
passing_reading_per_school = schools_students_complete_combined[schools_students_complete_combined['reading_score'] >= 50].groupby('school_name')['Student ID'].count() / students_per_school * 100

# Calculate the percentage of students passing both math and reading per school
passing_both_per_school = schools_students_complete_combined[(schools_students_complete_combined['maths_score'] >= 50) & (schools_students_complete_combined['reading_score'] >= 50)].groupby('school_name')['Student ID'].count() / students_per_school * 100

# Get the school types
school_types = schools_complete.set_index('school_name')['type']

# Create a DataFrame to hold the school overview
school_overview = pd.DataFrame({
    'School Type': school_types,
    'Total Students': students_per_school,
    'Total School Budget': total_budget_per_school,
    'Per Student Budget': per_student_budget_per_school,
    'Average Maths Score': average_maths_score_per_school,
    'Average Reading Score': average_reading_score_per_school,
    '% Passing Maths': passing_maths_per_school,
    '% Passing Reading': passing_reading_per_school,
    '% Overall Passing': passing_both_per_school
})

# Set the float formatting for monetary values
monetary_columns = ['Total School Budget', 'Per Student Budget']
school_overview[monetary_columns] = school_overview[monetary_columns].applymap('${:,.2f}'.format)

# Display the school overview DataFrame as a formatted table
display(school_overview)
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
# Top Performing Schools (By % Overall Passing)
# Sort the school overview DataFrame by % Overall Passing in descending order
top_schools = school_overview.sort_values('% Overall Passing', ascending=False)

# Save the top performing schools in a DataFrame called "top_schools"
top_schools = top_schools.head()

# Display the top five performing schools
display(top_schools)
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
# Bottom Performing Schools (By % Overall Passing)
# Sort the school overview DataFrame by % Overall Passing in ascending order
bottom_schools = school_overview.sort_values('% Overall Passing')

# Save the bottom performing schools in a DataFrame called "bottom_schools"
bottom_schools = bottom_schools.head()

# Display the bottom performing schools
display(bottom_schools)
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Maths Scores by Year
# Create pandas Series for each year level
maths_scores_9 = schools_students_complete_combined[schools_students_complete_combined['year'] == 9].groupby('school_name')['maths_score'].mean()
maths_scores_10 = schools_students_complete_combined[schools_students_complete_combined['year'] == 10].groupby('school_name')['maths_score'].mean()
maths_scores_11 = schools_students_complete_combined[schools_students_complete_combined['year'] == 11].groupby('school_name')['maths_score'].mean()
maths_scores_12 = schools_students_complete_combined[schools_students_complete_combined['year'] == 12].groupby('school_name')['maths_score'].mean()

# Combine the Series into a DataFrame
maths_scores_by_year = pd.DataFrame({
    'Year 9': maths_scores_9,
    'Year 10': maths_scores_10,
    'Year 11': maths_scores_11,
    'Year 12': maths_scores_12
})

# Apply cleaner formatting to the displayed data
maths_scores_by_year = maths_scores_by_year.applymap('{:.2f}'.format)

# Display the DataFrame
display(maths_scores_by_year)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Scores by School Spending
# Create bins and labels for spending ranges
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Categorize spending based on the bins using pd.cut
school_overview["Spending Ranges (Per Student)"] = pd.cut(school_overview["Per Student Budget"].str.replace('$', '').astype(float), spending_bins, labels=labels)

# Calculate mean scores per spending range
spending_maths_scores = school_overview.groupby(["Spending Ranges (Per Student)"])["Average Maths Score"].mean()
spending_reading_scores = school_overview.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_maths = school_overview.groupby(["Spending Ranges (Per Student)"])["% Passing Maths"].mean()
spending_passing_reading = school_overview.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_overview.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()

# Create the spending summary DataFrame
spending_summary = pd.DataFrame({
    "Average Maths Score": spending_maths_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Maths": spending_passing_maths,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

# Display the spending summary DataFrame
display(spending_summary)
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Scores by School Size
# Create bins and labels for school size
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Categorize school size based on the bins using pd.cut
school_overview["School Size"] = pd.cut(school_overview["Total Students"], size_bins, labels=labels)

# Create a DataFrame called size_summary
size_summary = pd.DataFrame({
    "Average Maths Score": school_overview.groupby("School Size")["Average Maths Score"].mean(),
    "Average Reading Score": school_overview.groupby("School Size")["Average Reading Score"].mean(),
    "% Passing Maths": school_overview.groupby("School Size")["% Passing Maths"].mean(),
    "% Passing Reading": school_overview.groupby("School Size")["% Passing Reading"].mean(),
    "% Overall Passing": school_overview.groupby("School Size")["% Overall Passing"].mean()
})

# Display the size_summary DataFrame
display(size_summary)
```
